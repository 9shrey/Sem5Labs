# Question 1

```python
import matplotlib.pyplot as plt
x = [0,100]
y = [0,200]


fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_yticks([0,50,100,150,200])
ax.plot(x, y)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')

plt.show()
```

# Question 2

```python
x = y

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])       
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) 

ax1.plot(x, y)
ax2.plot(x, y)

plt.show()
```

# Question 3

```python
import pandas as pd

data = pd.read_csv('company_sales_data.csv')

x = data['month_number']
y = data['total_profit']

plt.plot(x, y)

plt.xlabel('Month Number')
plt.ylabel('Total Profit')

plt.show()
```

# Question 4

```python
x = data['month_number']
y = data['total_units']

plt.plot(x, y, linestyle='--', color='red', marker='o', markerfacecolor='red', markersize=8, linewidth=3, label='Sold units')

plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')

plt.show()
```
